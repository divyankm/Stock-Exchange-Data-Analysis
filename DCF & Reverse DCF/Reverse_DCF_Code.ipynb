```python
"""
Reverse DCF Model Zerodha Varsity based on two Step CashFlows.This Chapter was introduced in 2017 on Fundamental Analysis Module,removed in 2022. 
Currently DCF Model is present Zerodha Varsity Website- https://zerodha.com/varsity/chapter/discounted-cash-flow-analysis-dcf/
Sharing PDF of Detailed DCF Working and Analysis.
Link-  https://einvestingforbeginners.com/reverse-dcf-daah/
"""

##User input
# symbol = ["KMC Hospital","INFY","TCS"]  ##Symbol is the name of company
"""
avg_cf is the average cashflow for past 3 consecutive years.
Suppose OA= Cash Generated from Operating Activities
        OF= Cash Generated From Financial Activities
        CE= Cash Consumed From Captial Expenditure (Fixed Assets Purchased)

For Example- FY20, Company XYZ has OA=50 Cr , OF=-45 Cr , CE= -20 Cr then Free Cash Flow (FCF) for FY 19 =OA + CE = 50 + (-20) = 30 Cr
Similarly FY 20 FCF = 25 Cr and FY21= 60 Cr then avg_cf will be , avg_cf = (FCF of FY19 + FCF of FY20 + FCF of FY21)/3 = (30 + 25 +60)/3 = 38.3 Cr

fcf_growth_for_first_5_years & fcf_growth_for_last_5_years can be changed dynamically based on Company.
dis_rate is considered between 8% to 10 %
years = 10 , means we are checking for 10 years off Future CashFlow
t_g_r is considered as Inflation percentage, which will vary in different countries. Current India's Long Term Inflation Rate is between 4% to 5% on 24 Sep 2022 as per RBI Stat's.
"""

fcf_growth_for_first_5_years = 14/100
fcf_growth_for_last_5_years = 7/100
dis_rate = 10/100
t_g_r = 4/100    ##terminal growth rate to be close to the long-term inflation number of the country, so between 4 and 5%.

"""
net_debt = 5.9   #Total Debt = Short Term Debt  + Long term Debt - Total cash in Hand
total_shares = 163085000  #Total Number of Shares of the company
cmp is the Current Market Price
mos_40 is the Margin of Safety, which we considered as 40% from the Calculated DCF Value of Stock or intrinsic value of stock.
"""
##Contant Terms
years = 10

!pip install yahoofinancials  #install libraries
!pip install sorcery
from yahoofinancials import YahooFinancials
import pandas as pd
from sorcery import dict_of
#https://pypi.org/project/yahoofinancials/

global total_shares
def change_to_cr(number):
    ##https://stackoverflow.com/questions/65041446/how-to-put-decimal-after-2-digits-in-a-string-in-python
    global total_shares
    if len(str(number))==8:
        n = str(number)
        n= n[:1] + '.' + n[1:]
        if number == total_shares:
            total_shares_in_cr=float(str(n)[:4])
            return total_shares_in_cr
        elif number == avg_cf:
            avg_cf_in_cr=float(str(n)[:4])
            i[1] = avg_cf_in_cr
            return avg_cf_in_cr
        elif number == net_debt:
            net_debt_in_cr=float(str(n)[:4])
            i[2] = net_debt_in_cr
            return net_debt_in_cr
        # return f'{number}_in_cr'
    elif len(str(number))==9:
        n = str(number)
        n= n[:2] + '.' + n[2:]
        if number == total_shares:
            total_shares_in_cr=float(str(n)[:4])
            return total_shares_in_cr
        elif number == avg_cf:
            avg_cf_in_cr=float(str(n)[:4])
            i[1] = avg_cf_in_cr
            return avg_cf_in_cr
        elif number == net_debt:
            net_debt_in_cr=float(str(n)[:4])
            i[2] = net_debt_in_cr
            return net_debt_in_cr
    elif len(str(number))==9:
        n = str(number)
        n= n[:2] + '.' + n[2:]
        if number == total_shares:
            total_shares_in_cr=float(str(n)[:4])
            return total_shares_in_cr
        elif number == avg_cf:
            avg_cf_in_cr=float(str(n)[:4])
            i[1] = avg_cf_in_cr
            return avg_cf_in_cr
        elif number == net_debt:
            net_debt_in_cr=float(str(n)[:4])
            i[2] = net_debt_in_cr
            return net_debt_in_cr
    elif len(str(number))==10:
        n = str(number)
        n= n[:3] + '.' + n[3:]
        if number == total_shares:
            total_shares_in_cr=float(str(n)[:4])
            return total_shares_in_cr
        elif number == avg_cf:
            avg_cf_in_cr=float(str(n)[:4])
            i[1] = avg_cf_in_cr
            return avg_cf_in_cr
        elif number == net_debt:
            net_debt_in_cr=float(str(n)[:4])
            i[2] = net_debt_in_cr
            return net_debt_in_cr
    elif len(str(number))==11:
        n = str(number)
        n= n[:4] + '.' + n[4:]
        if number == total_shares:
            total_shares_in_cr=float(str(n)[:4])
            return total_shares_in_cr
        elif number == avg_cf:
            avg_cf_in_cr=float(str(n)[:4])
            i[1] = avg_cf_in_cr
            return avg_cf_in_cr
        elif number == net_debt:
            net_debt_in_cr=float(str(n)[:4])
            i[2] = net_debt_in_cr
            return net_debt_in_cr

def dcf_cal(symbol,avg_cf,net_debt,total_shares):
    cf = pd.DataFrame()
    cf.insert(0, 'Sr_No', range(1,11))
    cf.insert(1, 'Year', range(23,33))

    ##fcf calculaton for 10 Years
    ##https://stackoverflow.com/questions/73440804/use-the-above-row-to-calculate-the-value-for-below-row-iteratively-using-pandas/73440941#73440941
    fcf=[]
    for row in range(len(cf)):
        if cf.Sr_No[row]==1:
            fcf.append(avg_cf*(1+fcf_growth_for_first_5_years))
        elif cf.Sr_No[row]<6:
            fcf.append(fcf[row-1]*(1+fcf_growth_for_first_5_years))
        else:
            fcf.append(fcf[row-1]*(1+fcf_growth_for_last_5_years)) 
    cf['fcf'] = fcf
    cf['pv'] = cf['fcf']/(1+dis_rate)**cf['Sr_No']
    ty = cf['Year'].tail(1).values[0]
    tv = float((cf['fcf'].tail(1)*(1+t_g_r)/(dis_rate - t_g_r)))
    pv_of_tv = (tv/(1+dis_rate)**10)
    total_pv_of_cf = cf['pv'].sum()+pv_of_tv
    price = round(((total_pv_of_cf - net_debt)*10**7)/total_shares)
    print(f"{symbol} total price:",price)
    print(f"{symbol} total Shares:",total_shares)
    print(f'Total Share in cr {symbol}',total_shares_in_cr)
    print(f'Net Debt in cr {symbol}',net_debt_in_cr)
    verify_price = round(int((total_pv_of_cf - net_debt)/total_shares_in_cr))
    print(f"Verify whether Reverse DCF Calculation are correct or not for {symbol} in Rs:",verify_price)
    print(f"Price in Rs for {symbol}:",price)
    mos_40 = (price - ((price/10)*4))
    print(f"Margin of Safety (40%) is Rs:{symbol}",mos_40)
    cmp = YahooFinancials(symbol)
    cmp = cmp.get_current_price()
    print(f"Current Market Price for {symbol:}",cmp)
    global d
    d = dict_of(symbol,avg_cf,fcf_growth_for_first_5_years,fcf_growth_for_last_5_years,years,t_g_r,dis_rate,net_debt,total_shares,total_shares,ty,tv,pv_of_tv,total_pv_of_cf,price,
                verify_price,mos_40,cmp)

# dcf_cal(symbol,avg_cf,net_debt,total_shares)
lst = [["KMCSHIL.BO",0,0,109107370],["PDMJEPAPER.NS",0,0,94950000]]
# fin_stm_all = pd.DataFrame() #https://stackoverflow.com/questions/28669482/appending-pandas-dataframes-generated-in-a-for-loop

for i in lst:
    l = str([i[0]]) 
    l = str(l)[1:-1]  #https://stackoverflow.com/questions/13207697/how-to-remove-square-brackets-from-list-in-python
    l = eval(l)       #https://stackoverflow.com/questions/40950791/remove-quotes-from-string-in-python
    l = str(l).replace("'", '"') #https://stackoverflow.com/questions/42183479/i-want-to-replace-single-quotes-with-double-quotes-in-a-list
    yahoo_financials = YahooFinancials(l)
    fin = yahoo_financials.get_financial_stmts('annual', ['income', 'cash', 'balance'])
    date_indexed_dict = {}
    for sub_dict in fin.values():
        for list_item in sub_dict[l]:
            for date, col_dict in list_item.items():
                if date not in date_indexed_dict:
                    date_indexed_dict[date] = {}
                # add the columns as keys on each index (date)
                for col, val in col_dict.items():
                    date_indexed_dict[date][col] = val

    # print(date_indexed_dict)
    fin_stm = pd.DataFrame.from_dict(date_indexed_dict, orient='index')
    fin_stm.index.names = ['Date']
    fin_stm = fin_stm.head(3)
    fin_stm['fcf'] = fin_stm['totalCashFromOperatingActivities'] - (- fin_stm['capitalExpenditures']) 
    fin_stm['net_debt'] = - (fin_stm['shortLongTermDebt'].head(1) + fin_stm['longTermDebt'].head(1)) + fin_stm['cash'] #Add - sign if value is positive
    avg_cf = int(fin_stm['fcf'].mean())
    net_debt = int(fin_stm['net_debt'].mean())
    print(f"Symbol {i[0]} Net Debt",net_debt)  
    print(f"Symbol {i[0]} Avg CF:",avg_cf)
    total_shares = i[3]
    print(f"Total shares for {i[0]}",total_shares)
    total_shares_in_cr = change_to_cr(total_shares)
    print(f"Total shares for {i[0]}",total_shares_in_cr)
    avg_cf_in_cr = change_to_cr(avg_cf)
    print(f"Avg CF for {i[0]}",avg_cf_in_cr)
    net_debt_in_cr = change_to_cr(net_debt)
    print(f"Net Debt for {i[0]}",net_debt_in_cr)
    # i.extend([avg_cf,net_debt])

pd.set_option('display.max_columns', None)
dicts = []
for item in lst:
    dcf_cal(*item)
    dicts.append(d)
dcf = pd.DataFrame(dicts)
dcf
```
